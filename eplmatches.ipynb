```python
# Importando bibliotecas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# leitura do arquivo
df = pd.read_csv("eplmatches.csv")
```

```python
df.shape
```

```python
# Exibindo as 5 primeiras linhas
df.head()
```

```python
# Verificando informações do Dataset: Entradas, Colunas e Tipos de dados
df.info()
```

```python
# Transfomando Date em tipo Datetime
df["Date"]=pd.to_datetime(df["Date"])
```

```python
# Verificando informações do Dataset: Entradas, Colunas e Tipos de dados
df.info()
```

```python
# Criando Coluna de Gols totais por partida
df["TotalGoals"] = df["HomeGoals"] + df["AwayGoals"]
```

```python
df["GoalsDifference"] = df["HomeGoals"] - df["AwayGoals"]
```

```python
# Exibindo as 5 primeiras linhas com nova coluna
df.head()
```

```python
# Verificando média de Gols
df["HomeGoals"].mean()
```

```python
# Verificando média de Gols
df["AwayGoals"].mean()
```

```python
# Verificando média de Gols
df["TotalGoals"].mean()
```

```python
# Verificando quantidade de gols por temporada

df.groupby("Season_End_Year")["TotalGoals"].sum()
```

```python
# Verificando quantidade de gols por temporada - Gráfico

df.groupby("Season_End_Year")["TotalGoals"].sum().plot.bar()
```

```python
# Verificando quantidade de jogos por temporada
df["Season_End_Year"].value_counts()
```

```python
# Verificando quantidade de jogos por temporada - Gráfico
df["Season_End_Year"].value_counts().plot.bar()
```

```python
# Média de gols por temporada
df.groupby("Season_End_Year")["TotalGoals"].mean()
```

```python
# Média de gols por temporada - Gráfico
df.groupby("Season_End_Year")["TotalGoals"].mean().plot.bar()
```

```python
# Média de gols feitos por time jogando em casa
df.groupby("Home")["HomeGoals"].mean().sort_values(ascending=False)
```

```python
# Média de gols feitos por time jogando Fora de casa
df.groupby("Away")["AwayGoals"].mean().sort_values(ascending=False)
```

```python
# Criando dataset só com a temporada de 2022

df_2022 = df[df["Season_End_Year"] == 2022]
```

```python
# Verificando Vitórias do time da casa, de fora e empates

df_2022["FTR"].value_counts()
```

```python
# Verificando time com maior média de gols em casa

df_2022.groupby("Home")["HomeGoals"].mean().sort_values(ascending=False)
```

```python
# Verificando time com maior média de gols fora casa

df_2022.groupby("Away")["AwayGoals"].mean().sort_values(ascending=False)
```
